# **Spacy summarisation**


```python
import spacy
nlp = spacy.load('en_core_web_sm')
doc = nlp(text)
```

```python
from spacy.lang.en.stop_words import STOP_WORDS
stopwords = list(STOP_WORDS)
```

```python
from string import punctuation
punctuation = punctuation + '\n'
```

```python
word_frequencies = {}
for word in doc:
  if word.text.lower() not in stopwords:
    if word.text.lower() not in punctuation:
      if word.text not in word_frequencies.keys():
        word_frequencies[word.text] = 1
      else:
        word_frequencies[word.text] += 1
```

```python
for word in word_frequencies.keys():
  word_frequencies[word] = word_frequencies[word]/max(word_frequencies.values())
word_frequencies
```

```python
sentence_tokens = [sent for sent in doc.sents]
sentence_tokens
```

```python
sentence_scores = {}
for sent in sentence_tokens:
  for word in sent:
    if word.text.lower() in word_frequencies.keys():
      if sent not in sentence_scores.keys():
        sentence_scores[sent] = word_frequencies[word.text.lower()]
      else:
        sentence_scores[sent] += word_frequencies[word.text.lower()]
sentence_scores
```

```python
from heapq import nlargest
select_length = int(len(sentence_tokens)*0.05)
summary = nlargest(select_length, sentence_scores, key = sentence_scores.get)
print(summary)
```

```python
final_summary = [word.text for word in summary]
print(final_summary)
```

```python
import pandas as pd
from youtube_transcript_api import YouTubeTranscriptApi as YTapi
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
def videoID(link):
    video_id = link.split("=")[1]
    return video_id
```

```python
def GetTranscript(video_id):
    try:
        transcript = YTapi.get_transcript(video_id)
        FinalTranscript = ' '.join([i['text'] for i in transcript])
    except Exception as e:
        print(e)

    return FinalTranscript
```

```python
checkpoint1 = "google/pegasus-large"
checkpoint2 = "csebuetnlp/mT5_multilingual_XLSum"
checkpoint3 = "sshleifer/distilbart-cnn-12-6"
checkpoint4 = "ai4bharat/IndicBART"
```

```python
tokenizer1 = AutoTokenizer.from_pretrained(checkpoint1)
model1 = AutoModelForSeq2SeqLM.from_pretrained(checkpoint1)
```

```python
tokenizer3 = AutoTokenizer.from_pretrained(checkpoint3)
model3 = AutoModelForSeq2SeqLM.from_pretrained(checkpoint3)
```

```python
tokenizer4 = AutoTokenizer.from_pretrained(checkpoint4)
model4 = AutoModelForSeq2SeqLM.from_pretrained(checkpoint4)
```

```python
def summarize(toenizer, model, text):

    """
    Function to summarize the text(transcript)
    """
    inputs = toenizer(text,
                    max_length=1024,
                    truncation=True,
                    return_tensors="pt")

    summary_ids = model.generate(inputs["input_ids"])
    summary = toenizer.batch_decode(summary_ids,
                                  skip_special_tokens=True,
                                  clean_up_tokenization_spaces=False)
    return summary
```

```python
pegasus = summarize(tokenizer1, model1, transcript_en)
```

# **By manually inputting transcripts**

```python
from transformers import pipeline

def summarize_subtitles(subtitles):
    summarizer = pipeline("summarization")
    summary = summarizer(subtitles, max_length=150, min_length=30, do_sample=False)
    return summary[0]['summary_text']

if __name__ == "__main__":
    # Example subtitles string
    subtitles = """
    In a land of myth, and a time of magic, the destiny of a great kingdom rests on the shoulders of a young boy. His name... Merlin. To some, magic is a way of life. To others, it is a forbidden art. Yet to Merlin, it is the very thing that will save his kingdom. When Merlin arrives in the city of Camelot, he encounters a kingdom where magic is outlawed and punishable by death. Despite the risks, Merlin must embrace his destiny and protect the young prince, Arthur, who will one day become the legendary King Arthur. Alongside his friends, the loyal Gwen, the wise Gaius, and the brave knights of the Round Table, Merlin must navigate the treacherous waters of court intrigue and dark magic to fulfill his destiny and unite the kingdom.
    """

    summary = summarize_subtitles(subtitles)
    print("Summary:")
    print(summary)
```
